### Just messing around with linear regression, seeing what I can do with it with the keras standard interface. 

```python
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plot

def generate_data(x):
    return np.sin(x) + np.random.uniform(-0.5, +0.5, x.shape)
```

```python
x = np.random.uniform(0,6,1000)
```

```python
y = generate_data(x)
plot.scatter(x,y)
plot.show()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation = tf.keras.activations.tanh, input_shape =(1,)),
    #tf.keras.layers.Dense(10, activation = tf.keras.activations.tanh),
    tf.keras.layers.Dense(1, activation = tf.keras.activations.linear)
])
```

```python
model.compile(optimizer = tf.train.AdamOptimizer(),
             loss = "mse",
             metrics = ['accuracy'])
```

```python
model.fit(x,y,epochs = 200, batch_size = 32, verbose = 0)
```

```python
plot.scatter(x,y)
plot.scatter(x, model.predict(x))
plot.show()
```
